```python
import random
import matplotlib.pyplot as plt

# Simple quote themes
quotes = {
    "fire": [
        "From ashes, something always rises.",
        "Burn quietly, but burn bright."
    ],
    "moon": [
        "Even shadows belong to the moon.",
        "She glows without needing permission."
    ],
    "hope": [
        "Hope is a stubborn flame.",
        "Even cracked glass reflects light."
    ],
    "default": [
        "Every word carries a spark.",
        "Stories live longer than moments."
    ]
}

# Random color palettes
colors = ["#1a1a1a", "#003049", "#2b2d42", "#000000", "#1d3557"]
text_colors = ["#f1faee", "#eae2b7", "#edf2f4", "#fffcf2", "#f8f9fa"]

def get_quote(keywords):
    for word in keywords:
        if word in quotes:
            return random.choice(quotes[word])
    return random.choice(quotes["default"])

def generate_poster(keywords):
    title = " ".join(word.capitalize() for word in keywords)
    quote = get_quote(keywords)

    bg = random.choice(colors)
    fg = random.choice(text_colors)

    # Plot with matplotlib
    fig, ax = plt.subplots(figsize=(8, 5))
    ax.set_facecolor(bg)
    plt.axis('off')

    plt.text(0.5, 0.7, title,
             fontsize=32, ha='center', va='center', color=fg, weight='bold')

    plt.text(0.5, 0.4, f'"{quote}"',
             fontsize=14, ha='center', va='center', color=fg, style='italic')

    plt.title("ðŸŽž Minimalist Story Poster", color=fg, fontsize=10)
    plt.show()

# ðŸ§  Input and Run
user_input = input("Enter 3 words that inspire your story (comma-separated):\n")
keywords = [w.strip().lower() for w in user_input.split(",") if w.strip()]

generate_poster(keywords)
```
